# 4. Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df.columns
```

```python
X = df.drop(columns=["Attrition", "AttritionEnc"])
y = df["AttritionEnc"]
```

```python
# One-Hot Encoding
X = pd.get_dummies(X, drop_first=True)
```

```python
X.columns
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
# Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train_scaled, y_train)

# Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)
```

```python
df["PerformanceRating"]
```

```python
df["PerformanceRating"].value_counts()
```

```python
# Predicting Performance Rating
y2 = df["PerformanceRating"]
X2 = df.drop(columns=["Attrition", "AttritionEnc", "PerformanceRating"])
X2 = pd.get_dummies(X2, drop_first=True)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, random_state=42, test_size=0.2)
X2_train_s = scaler.fit_transform(X2_train)
X2_test_s = scaler.transform(X2_test)
```

```python
from sklearn.metrics import classification_report

rf2 = RandomForestClassifier(n_estimators=100, random_state=42)
rf2.fit(X2_train_s, y2_train)
rf2_pred = rf2.predict(X2_test_s)
print(classification_report(y2_test, rf2_pred))
```

# 5. Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
```

```python
# Model evaluation function
def evaluate_model(model_name, y_test, y_pred, y_prob):
    print(f"{model_name} Evaluation:")
    print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(y_test, y_pred):.4f}")
    print(f"Recall:    {recall_score(y_test, y_pred):.4f}")
    print(f"F1-Score:  {f1_score(y_test, y_pred):.4f}")
    print(f"AUC-ROC:   {roc_auc_score(y_test, y_prob):.4f}\n")

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Stayed (0)', 'Left (1)'],
                yticklabels=['Stayed (0)', 'Left (1)'])
    plt.title(f'Confusion Matrix: {model_name}')
    plt.ylabel('Actual Reality')
    plt.xlabel('Model Prediction')
    plt.show()
```

```python
# Prediction and Probability for Logistic Regression
lr_pred = lr.predict(X_test_scaled)
lr_prob = lr.predict_proba(X_test_scaled)[:, 1]
```

```python
# Prediction and Probability for Random Forest Classifier
rf_pred = rf.predict(X_test_scaled)
rf_prob = rf.predict_proba(X_test_scaled)[:, 1]
```

```python
# Evaluation
evaluate_model("Logistic Regression", y_test, lr_pred, lr_prob)
evaluate_model("Random Forest", y_test, rf_pred, rf_prob)
```

```python
# Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5]
}
grid = GridSearchCV(RandomForestClassifier(random_state=42, class_weight="balanced"), param_grid, cv=5, scoring="recall")
grid.fit(X_train_scaled, y_train)
best_rf = grid.best_estimator_
print(grid.best_params_)
```

```python
best_prob = best_rf.predict_proba(X_test_scaled)[:, 1]
# best_pred = (best_prob >= 0.5).astype(int)
best_pred = (best_prob >= 0.25).astype(int)
evaluate_model("Random Forest (Tuned)", y_test, best_pred, best_prob)
```

```python
# Prediction and Probability
perf_pred = perf_model.predict(X_perf_test)
perf_prob = perf_model.predict_proba(X_perf_test)[:, 1]
```

```python
# Evaluation
evaluate_model("Performance Rating (Decision Tree)", y_perf_test, perf_pred, perf_prob)
```
